```python
import datetime
import numpy as np
import pandas as pd
import pandas.io.data as web

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('ticks')

import elhers_indicators as ei
import series_generator as sg
from exponential_MA import EMA
```

##Retrieve data

```python
start = datetime.datetime(2015, 1, 1)

end = datetime.datetime(2015, 8, 30)

aapl = web.DataReader("AAPL", 'yahoo', start, end)

series = aapl.ix[:,"Adj Close"].as_matrix()
idx = range(series.shape[0])
```

```python
EMA_5 = EMA(series, 5)
EMA_15 = EMA(series, 15)
EMA_30 = EMA(series, 30)

SS_5 = ei.super_smoother(series, 5)
SS_15 = ei.super_smoother(series, 15)
SS_30 = ei.super_smoother(series, 30)

decycle_5 = ei.decycle(series, 5)
decycle_15 = ei.decycle(series, 15)
decycle_30 = ei.decycle(series, 30)

IT_5 = ei.instantaneous_trend(series, 5)
IT_15 = ei.instantaneous_trend(series, 15)
IT_30 = ei.instantaneous_trend(series, 30)
```

##Comparative different indicator same period.

```python
plt.plot(idx, series, label="Series")
plt.plot(idx, EMA_5, label="EMA(5)")
plt.plot(idx, SS_5, label="SuperSmoother(5)")
plt.plot(idx, decycle_5, label="Decycle(5)")
plt.plot(idx, IT_5, label="InstantaneousTrend(5)")
plt.legend(loc="lower right")
plt.grid()
```

```python
plt.plot(idx, series, label="Series")
plt.plot(idx, EMA_15, label="EMA(15)")
plt.plot(idx, SS_15, label="SuperSmoother(15)")
plt.plot(idx, decycle_15, label="Decycle(15)")
plt.plot(idx, IT_15, label="InstantaneousTrend(15)")
plt.legend(loc="lower right")
plt.grid()
```

```python
plt.plot(idx, series, label="Series")
plt.plot(idx, EMA_30, label="EMA(30)")
plt.plot(idx, SS_30, label="SuperSmoother(30)")
plt.plot(idx, decycle_30, label="Decycle(30)")
plt.plot(idx, IT_30, label="InstantaneousTrend(30)")
plt.legend(loc="lower right")
plt.grid()
```
